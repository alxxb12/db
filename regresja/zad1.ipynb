```python
import numpy as np

import statsmodels.api as sm
import seaborn as sea
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import pandas as pd
import plotnine as p9
```

```python
data=pd.read_csv("exercise.csv")
exc1 = pd.DataFrame(data,columns=['y','x1','x2'])[:40]
exc1
```

```python

est = smf.ols(formula='y ~ x1 + x2', data=exc1).fit()
est.summary()
```

```python
figure1 = (p9.ggplot(p9.aes(x='x1', y='y'), data=exc1) 
        + p9.geom_jitter(width=0.1))
figure1
```

```python
figure2 = (p9.ggplot(p9.aes(x='x2', y='y'), data=exc1) 
        + p9.geom_jitter(width=0.1))
figure2
```

```python
sea.residplot(exc1['x1'], exc1['y'], lowess=True)
```

```python
sea.residplot(exc1['x2'], exc1['y'], lowess=True)
```
