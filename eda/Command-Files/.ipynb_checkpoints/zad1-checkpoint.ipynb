```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
import numpy as np
```

```python
df = pd.read_csv("../Original-Data/tb.csv")
```

```python
df
```

```python
def rangeage(ager):
    if(len(ager)==4):
        
        return ager[:2]+"--"+ager[2:]
    
    elif(len(ager)==3):
        
        return ager[:1]+"--"+ager[1:]
    
    else:
        return ager+"+"
```

```python
df =df.rename(columns={'iso2':'country'})
df =df.rename(columns=lambda x: re.sub('new_sp_','',x))
df =df.drop(['new_sp', 'mu', 'fu', 'm04', 'f04', 'm514', 'f04', 'f514'], axis=1)
```

```python
df=pd.melt(df,id_vars=['country', 'year'], var_name='column', value_name='cases')
df['sex']=df['column'].str[0]
df['age']=df["column"].str[1:].apply(rangeage)
df=df.drop(['column'],axis=1)
df=df[np.isfinite(df['cases'])]
df['cases']=df['cases'].apply(lambda x: int(x))
```

```python
df
```

Tb Dataset zawiera dane o gruźlicy w różnych grupach pacjentów z podzialem na plec, przedzial wiekowy , rok czy lokalizacje

Podział na plec:

```python
sex_counts = df.groupby('sex')['cases'].agg('sum')
```

sns.barplot(x=sex_counts.index, y=sex_counts.values)

Podział na przedział wiekowy

```python
age_counts = df.groupby('age')['cases'].agg('sum')
sns.barplot(x=age_counts.index, y=age_counts.values)
```

Podzial na lokacje gdzie wystepowanie bylo najwieksze

```python
location_counts = df.groupby('country')['cases'].agg('sum')
sorted_location_counts = location_counts.sort_values()
location = sorted_location_counts[-5::]
sns.barplot(x=location.index, y=location.values)
```
